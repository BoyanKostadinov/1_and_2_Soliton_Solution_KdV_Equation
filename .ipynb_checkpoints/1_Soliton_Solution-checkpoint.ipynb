# Derivation of One-Soliton Solution to the Korteweg-De Vries Equation using the Hirota Method

```python
import sympy as sp
```

```python
# Define variable
x, t, k, c = sp.symbols('x t k c', real=True)
```

```python
# Define functions
u = sp.Function('u')(x, t)
f = sp.Function('f')(x, t)
```

$$
u_t + 6 u u_x + u_{xxx} = 0
\tag{KdVEq}
$$

Here $u = u(x, t)$ is a function of $x$ and $t$.

In the present document we consider the Korteweg-De Vries equation (KdVEq)

```python
# Define the Korteweg-De Vries equation (KdVEq)
kdv_eq = sp.diff(u, t) + 6 * u * sp.diff(u, x) + sp.diff(u, x, 3)
```
